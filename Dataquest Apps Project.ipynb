This project focuses on the analysis of free mobile data usage, with the aim of helping app developers understand which types of applications are more likely to attract users on Google Play and the App Store.


In this project we will work with two dataset. The first one conteins informations about applications sourced from AppStore, here is the link to connect to the repoistory where the data are located: [Mobile App Store](https://www.kaggle.com/datasets/ramamet4/app-store-apple-data-set-10k-apps).
```python
from csv import reader

opened_file = open('AppleStore.csv', encoding='utf-8')
readed_file = reader(opened_file)
ios = list(readed_file)
ios_header = ios[0]
ios = ios[1:]

opened_file = open('googleplaystore.csv', encoding='utf-8')
readed_file = reader(opened_file)
android = list(readed_file)
android_header = android[0]
android = android[1:]
```

```python
def explore_data(dataset, start, end, row_column = False):
    dataset_sliced = dataset[start:end]
    for row in dataset_sliced:
        print(row)
        print('\n')
    if row_column:
        print('The number of rows is: ', len(dataset))
        print('\n')
        print('The number of columns is: ', len(dataset[0]))
print(ios_header)
print('\n')
explore_data(ios, 0, 3, True)
```

`user_rating` measures the level of satisfaction that users assign to an application by giving it a score from 0 to 5. Applications with higher values tend to be used more frequently.

Finally, `prime_genre`e indicates the category to which applications belong. This factor can be important for understanding which genres are generally preferred.

## Removing duplicated rows
Another frequent type of error is that of duplicate rows.

```python
for element in android:
    name = element[0]
    if name == 'Instagram':
        print(element)
        print('\n')
```

We can see that Instagram is reported more than one time. The only value that change for each row is in the fourth column: `Reviews`. Instead of deleting the repeted rows randomly, it's wiser to leave the row with the most reviews because that means is the rows with the most updated informations. 
But for now, let's write a code that divides duplicated rows from the unique ones.

```python
duplicated_apps = []
unique_apps = []
for rows in android:
    name = rows[0]
    if name in unique_apps:
        duplicated_apps.append(name)
    else:
        unique_apps.append(name)
print('The number of duplicated rows is:',len(duplicated_apps))
```

At this point, among the duplicate rows, I keep only the one that reports the highest number of reviews. This choice is made because the row with the highest number of reviews is the most up to date.

```python
reviews_max = {}

for app in android:
    name = app[0]
    n_reviews = float(app[3])
    
    if name in reviews_max and reviews_max[name] < n_reviews:
        reviews_max[name] = n_reviews
        
    elif name not in reviews_max:
        reviews_max[name] = n_reviews
```

```python
print('Expected lenght:', len(android) - 1181)
print('Actual lenght:', len(reviews_max))
```

Having done this, I clean the initial dataset so that in the next steps I can work with a version that does not contain duplicate rows.

```python
android_clean = []
already_added = []
for app in android:
    name = app[0]
    n_reviews = float(app[3])
    if (reviews_max[name] == n_reviews) and (name not in already_added):
        android_clean.append(app) 
        already_added.append(name)
```

```python
print(len(android_clean))
```

## Removing the non-english apps
The initial objective is to gather information about the characteristics of the most widely used applications in order to focus time and resources on the development of free applications whose only source of revenue comes from in-app advertising. The target audience for future applications is English-speaking users; therefore, it is important to remove from the datasets applications that are in foreign languages.

```python
def english_apps(string):
    for character in string:
        if ord(character) > 127:
            return False
        
    return True
print(english_apps('Instagram'))
print(english_apps('Áà±Â•áËâ∫PPS -„ÄäÊ¨¢‰πêÈ¢Ç2„ÄãÁîµËßÜÂâßÁÉ≠Êí≠'))
print(english_apps('Docs To Go‚Ñ¢ Free Office Suite'))
print(english_apps('Instachat üòú'))
```

The problem here is that applications such as `Docs To Go‚Ñ¢ Free Office Suite` or `Instachat üòú` contain symbols that are not among the first 127 ASCII characters. In order not to exclude valuable information from the analysis, we need to find another method to determine whether an app belongs to the ‚ÄúEnglish‚Äù or ‚ÄúNot English‚Äù category.

```python
def is_english(string):
    non_ascii = 0

    for characters in string:
        if ord(characters) > 127:
            non_ascii += 1
    if non_ascii > 3:
        return False
    else:
        return True
print(is_english('Instagram'))
print(is_english('Áà±Â•áËâ∫PPS -„ÄäÊ¨¢‰πêÈ¢Ç2„ÄãÁîµËßÜÂâßÁÉ≠Êí≠'))
print(is_english('Docs To Go‚Ñ¢ Free Office Suite'))
print(is_english('Instachat üòú'))
```

Once the seemingly best method for identifying foreign-language applications has been found, we apply the code above to both of our datasets.

```python

english_android = []
english_ios = []

for rows in android_clean:
    name = rows[0]
    if is_english(name):
        english_android.append(rows)
for rows in ios:
    name = rows[0]
    if is_english(name):
        english_ios.append(rows)
explore_data(english_android, 0, 3, True)
print('\n')
explore_data(english_ios, 0, 3, True)
    
        
   
```

# Most common apps by genre
Because our end goal is to add the app on both Google Play and the App Store, we need to find app profiles that are successful in both markets. For instance, a profile that works well for both markets might be a productivity app that makes use of gamification.

Using the following code, a frequency table was created to illustrate, for each application genre, the distribution of apps in terms of percentage. This allows us to see what proportion of the total applications belongs to each genre, providing a clear overview of the popularity of different types of apps.

```python
def freq_table(dataset, index):
    table = {}
    total = 0
    
    for row in dataset:
        total += 1
        value = row[index]
        if value in table:
            table[value] += 1
        else:
            table[value] = 1
    
    table_percentages = {}
    for key in table:
        percentage = (table[key] / total) * 100
        table_percentages[key] = percentage 
    
    return table_percentages


def display_table(dataset, index):
    table = freq_table(dataset, index)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)
        
    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', entry[0])
```

```python
display_table(ios_final, -5)
```

## Most popular apps by Genre in AppStore
One way to determine which app genres are the most popular (i.e., have the largest user base) is to calculate the average number of installs for each genre. In the Google Play dataset, this information is available in the `Installs` column. However, for the App Store dataset, this data is not provided. As a workaround, we can use the total number of user ratings, available in the rating_count_tot column, as a proxy for popularity.

Below, we calculate the average number of user ratings per app genre in the App Store:

```python
genres_ios = freq_table(ios_final, -5)

for genre in genres_ios:
    total = 0
    len_genre = 0
    for app in ios_final:
        genre_app = app[-5]
        if genre_app == genre:            
            n_ratings = float(app[5])
            total += n_ratings
            len_genre += 1
    avg_n_ratings = total / len_genre
    print(genre, ':', avg_n_ratings)
```

Apparentemente, le applicazioni che appartengono al genere "Reference" sono quelle che contano maggiori recensioni e quindi, plausibilmente, sono quelle che appartengono alla categoria maggiormente installata nei dispositivi. Rientrano in questo genere, dizionari, enciclopedie, guide, mappe, app di consultazione tecnica o culturale.

```python
category_ft = freq_table(android_final, 1)

for category in category_ft:
    total = 0
    len_category = 0

    for element in android_final:
        category_app = element[1]

        if category_app == category:
            n_install = element[5]
            n_install = n_install.replace(',', '')
            n_install = n_install.replace('+', '')
            
            total += float(n_install)
            len_category += 1
            
    avg_n_install = total / len_category
    print(category, ':', avg_n_install)
```
